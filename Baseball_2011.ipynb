```python
scores = baseball_2011_small[['GAME_ID','HOME_TEAM_ID','AWAY_TEAM_ID','INN_CT','HOME_SCORE_CT','AWAY_SCORE_CT', ]].copy()
```

```python
#gets the final score of the games, get last line of each game
scores_final = scores.groupby('GAME_ID').last()
scores_final.head()
```

```python
#check for tied games
scores_final['DRAWS'] = scores_final['HOME_SCORE_CT'] == scores_final['AWAY_SCORE_CT']
```

```python
#there are no ties in baseball, add rbi and batting team columns to help calculate final scores
scores = baseball_2011_small[['GAME_ID','HOME_TEAM_ID','AWAY_TEAM_ID','INN_CT','BAT_HOME_ID','RBI_CT','HOME_SCORE_CT','AWAY_SCORE_CT']].copy()
```

```python
scores.head()
```

```python
scores_final = scores.groupby('GAME_ID').last()
scores_final.head()
```

```python
#calculate final score, add rbi to team at bat, 0=away, 1=home
for team in scores_final['BAT_HOME_ID']:
    if team == 1:
        scores_final['HOME_SCORE_FN'] = scores_final['HOME_SCORE_CT']+scores_final['RBI_CT']
        scores_final['AWAY_SCORE_FN'] = scores_final['AWAY_SCORE_CT']
```
