```python
def state_list():
    bases = ['0', '1', '2', '3', '12', '13', '23', '123']
    outs = ['', 'X', 'XX', 'XXX']
    states = []
    for b in bases:
        for x in outs:
            states.append(b+x)
    return(states)

def transient_state_list():
    bases = ['0', '1', '2', '3', '12', '13', '23', '123']
    outs = ['', 'X', 'XX', 'XXX']
    t_states = []
    for b in bases:
        for x in outs[0:3]:
            t_states.append(b+x)
    return(t_states)

def returning_transitions():
    t_states = transient_state_list()
    r_trans = []
    for s in t_states:
        r_trans.append(s+":"+s)
    return(r_trans)
```

```python
def project(d):
    r_trans = returning_transitions()
    r_columns = list(set(r_trans).intersection(set(d.columns)))
    projected = d.drop(r_columns, axis=1)
    dropped = d[r_columns]
    return projected, dropped
```

```python
def all_transitions_from(a_transient_state, df):
    trans = []
    for s in states:
        trans.append(a_transient_state+':'+s)
    cols = list(set(df.columns).intersection(set(trans)))
    return(cols)
```

```python
def embed(p):
    states = state_list()
    t_states = transient_state_list()
    # Get list of a transitions from each state
    e = p.copy()
    e['dummy'] = -1
    for ts in t_states:
        ts_cols = all_transitions_from(ts, p)
        ts_cols.append('dummy')
        e[ts+':'+ts] = -e[ts_cols].sum(axis=1)
    e.drop('dummy', axis=1, inplace=True)
    return(e)
```
